```python
#import important packages and libraries
import cv2
import easyocr
import numpy as np
from numpy import ndarray
from google.colab.patches import cv2_imshow
```

```python
#reading image
image_path = '/content/hand2.jpeg'
img = cv2.imread(image_path)
cv2_imshow(img)
```

```python
#instant text reader
picture_reader = easyocr.Reader(['en'], gpu=False)
```

```python
#text detection on image
picture_result = picture_reader.readtext(image_path)
for t in picture_result:
  print(t)
```

```python
#draw a bounding box on image
for y in picture_result:
  top_left_corner = [int(value) for value in y[0][0]]
  bottom_right_corner = [int(value) for value in y[0][2]]
  text = y[1]
  img = cv2.rectangle(img, top_left_corner, bottom_right_corner, (0, 0, 255), 2)
  img = cv2.putText(img, text, top_left_corner, cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
cv2_imshow(img)
```
